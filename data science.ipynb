```python
x = [15,82,33,6,95] 
 
y = [27,48,69,32,0] 

plt.plot(x,y) 

plt.show()
```

```python
x = [1,2,3] 
y = [2,4,6] 

plt.plot(x, y, label = "line1") 

x1 = [1,2,3] 
y1 = [4,8,3] 

plt.plot(x1, y1, label = "line2") 

plt.xlabel('x-axis') 

plt.ylabel('y-axis') 

plt.title('Two lines on one graph') 

plt.legend() 

plt.show()
```

```python
x = [2,4,6,8,10,12] 

y = [6,9,1,5,2,6] 


plt.plot(x, y, color='pink', linestyle='dashed', linewidth = 3, marker='o', markerfacecolor='yellow', markersize=12) 

plt.ylim(1,8) 
plt.xlim(1,8) 

plt.xlabel('x-axis') 

plt.ylabel('y-axis') 

plt.show()
```

```python
food = ['Burgerking', 'Mcdonalds', 'KFC', 'TexasChicken'] 

s = [6, 10, 8, 6] 

colors = ['r', 'y', 'g', 'b'] 

plt.pie(s, labels = food, colors=colors, startangle=90, shadow = True, explode = (0.2, 0, 0, 0), radius = 1.2, autopct = '%1.1f%%') 

plt.legend() 

plt.show()
```

```python
x = np.arange(0, 10*(np.pi), 0.52) 
y = np.sin(x) 

plt.plot(x, y) 

plt.show()
```

```python
xlist = np.linspace(-10.0, 10.0, 100)
ylist = np.linspace(-5.0,53.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) 
ax.set_title('Filled Contours Plot')
ax.set_ylabel('y (cm)')
plt.show()
```
